# Contours

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
def img_show(img):
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def contours(path):
    img = cv2.imread(path)
    imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(imgray, 127, 255, 0)
    im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(img, contours, -1, (0, 0, 255), 3)
    img_show(img)
```

```python
contours('datasets/contour.png')
```

```python
contours('datasets/simulating_protanopia.png')
```
