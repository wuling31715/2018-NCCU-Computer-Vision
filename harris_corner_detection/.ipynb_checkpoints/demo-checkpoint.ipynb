# Harris Corner Detector

### Theory

$ 
E(u,v) =  
  \left[
    \begin{array}{rrr}
      u & v \\
    \end{array}
  \right]
  M
    \left[
    \begin{array}{rrr}
      u  \\
      v  \\
    \end{array}
  \right]
$

$
M = \sum W(x,y)
  \left[
    \begin{array}{rrr}
      I_xI_x & I_xI_y \\
      I_yI_x & I_yI_y \\
    \end{array}
  \right]
$

* $ R = det(M) - k(trace(M))^2 $

* $ det(M) = \lambda1 * \lambda2 $

* $ trace(M) = \lambda1 + \lambda2 $

* $ \lambda1, \lambda2 $ are the eigen values of M

### Implement

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def harris_corner_detector(path):
    img = cv2.imread(path)
    origin = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = np.float32(gray)
    dst = cv2.cornerHarris(gray,10,3,0.04)
    dst = cv2.dilate(dst,None)
    img[dst>0.01*dst.max()] = [0,0,255]
    # Original
    plt.subplot(121)
    plt.imshow(origin, cmap = 'rainbow')
    plt.title('Original')
    plt.xticks([])
    plt.yticks([])
    # Harris Corner Detector
    plt.subplot(122)
    plt.imshow(img, cmap = 'gray')
    plt.title('Harris Corner Detector')
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
harris_corner_detector('datasets/simulating_protanopia.png')
```
